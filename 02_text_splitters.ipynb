```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
```

```python
from langchain_community.document_loaders import PyPDFLoader

pdf_loader = PyPDFLoader("neet_2025.pdf")
docs = pdf_loader.load()
```

```python
neet_2025_paper_docs = text_splitter.split_documents(docs)
```

```python
type(neet_2025_paper_docs), len(neet_2025_paper_docs) # so basically we got 297 documents from 1 document
```

```python
type(neet_2025_paper_docs[0])
```

```python
print(neet_2025_paper_docs[0].page_content)
```

# Creating docs from text directly

```python
speech = ""

with open("speech.txt") as f:
    speech = f.read()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap= 50)
```

```python
speech_docs = text_splitter.create_documents([speech])
speech_docs
```

```python
type(speech_docs), len(speech_docs), type(speech_docs[0])
```

```python
print(speech_docs[0])
```

# HTMlHeaderTextSplitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on = [("h1", "Header 1"), ("h2", "Header 2")])
```

```python
html = """<html>
    <body>
        <h1>LangChain Framework</h1>
        <p>LangChain is a tool for building LLM apps.</p>
        
        <h2>Key Components</h2>
        <p>One of the main components is text splitters.</p>
        
        <h2>Pricing</h2>
        <p>LangChain is completely open-source and free.</p>
    </body>
</html>"""
```

```python
html_docs = html_splitter.split_text(html)
```

```python
html_docs
```

# Splitting HTML from url

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_text_splitter = HTMLHeaderTextSplitter(headers_to_split_on= [
    ("h1", "Header 1"), ("h2", "Header 2"), ("h3", "Header 3")
])
```

```python
scraped_docs = html_text_splitter.split_text_from_url("https://books.toscrape.com/")
```
