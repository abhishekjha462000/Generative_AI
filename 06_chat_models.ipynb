# Chat Models

Chat Models are used for the conversational tasks and they have their base class as BaseChatModel.

```python
from langchain_openai import ChatOpenAI

chat_model = ChatOpenAI(model = "gpt-3.5-turbo", temperature = 0)

result = chat_model.invoke("What is love in short")

```

```python
type(result)
```

```python
print(result.content)
```
