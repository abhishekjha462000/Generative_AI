```python
from langchain_community.document_loaders import TextLoader
```

```python
loader = TextLoader("speech.txt")
loader.load()
# The entire content inside the speech.txt will be converted into documents.
# Thus, using loader.load() will return to me a list of documents.
```

# Reading a pdf file

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
pdf_loader = PyPDFLoader(file_path="neet_2025.pdf")
```

```python
pdf_docs = pdf_loader.load()
```

```python
type(pdf_docs)
```

```python
dict(pdf_docs[0])
```

### whenever we try to print a doc in the kernel, automatically its page_content would be printed

```python
print(pdf_docs[0])
```

# Arxiv Loader

```python
from langchain_community.document_loaders import ArxivLoader
```

```python
arxiv_loader = ArxivLoader("1706.03762", load_max_docs = 2)
```

```python
arxiv_docs = arxiv_loader.load()
```

```python
arxiv_docs
```

```python
len(arxiv_docs)
```

```python
print(arxiv_docs[0].page_content)
```
