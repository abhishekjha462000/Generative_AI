# Prompts

Prompts are one of the main components of the langchain core library and they are of 2 types.

The first type is the static prompts and the other type is the dynamic prompts. The static prompts are the one which are fixed and can be directly passed into the llm during invokation.

The dynamic prompts are however created using the PromptTemplate which takes input variables and returns one massive, single string.

```python
from langchain_core.prompts import PromptTemplate

# LangChain automatically scans the string, sees {concept}, 
# and sets input_variables=["concept"] internally.
prompt = PromptTemplate.from_template(
    "Explain the math concept of {concept} in short."
)

prompt.input_variables
```

If someone look at tutorials from a year ago, you will see that people dont use the .from_template() method. Instead of using .from_template(), you instantiate the class directly. When you do this, LangChain forces us to pass in the list of input variables as well.

```python
from langchain_core.prompts import PromptTemplate

# This will STILL crash because you explicitly provided a list 
# that does not match the string. The validator catches the mismatch.
template = PromptTemplate(
    template="Explain the math concept of {concept} in short.",
    input_variables=["concecpt"] # <-- Typo here triggers the ValueError
)

template.input_variables
```

#### I take my words back. It's a legacy parameter — in older LangChain, auto-inference didn't exist, so you had to declare variables manually. It's kept for backward compatibility and for cases where you want explicit validation (e.g., ensuring certain variables are always present).

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template("Explain the math concept of {concept} in short.")

prompt = template.invoke({"concept": "Linear Algebra"})
prompt
```

```python
from langchain_openai import OpenAI

llm = OpenAI()

result = llm.invoke(prompt)

print(result)
```
